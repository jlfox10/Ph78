```python
import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.signal import chirp, spectrogram


from scipy import signal
import scipy

from scipy.fft import fft, fftshift

import csv

```

```python
X = []
Y = []

with open('dxdz.txt', 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=',')
    for ROWS in plotting:
        X.append(ROWS[-1])
        Y.append(ROWS[0])
    
    
plt.plot(X, Y)
plt.title('Line Graph using CSV')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x = []
y = []

with open('Bx1.csv', 'r') as file:
    csv_reader = csv.reader(file)
    plots = csv.reader(file, delimiter = ',')
    for _ in range(8):  # Skip first 3 rows
        next(csv_reader)

    for row in plots:
        x.append(float(row[0]))
        y.append(float(row[1]))
    
    
plt.plot(x, y)
plt.title('Perpendicular Magnetic Field 1')
plt.xlabel('Positition')
plt.ylabel('Field Strength (T)')
plt.show()
```
