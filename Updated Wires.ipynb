```python
import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.signal import chirp, spectrogram


from scipy import signal
import scipy

from scipy.fft import fft, fftshift

import csv

```

```python
x = []
y = []

with open('editedbx.csv', 'r') as file:
    csv_reader = csv.reader(file)
    plots = csv.reader(file, delimiter = ',')
    for _ in range(8):  # Skip first 3 rows
        next(csv_reader)

    for row in plots:
        x.append(float(row[0]))
        y.append(float(row[1]))

xap = []
yap = []
num = 0

for i in x:
    xap.append(i)
    ind = x.index(i)
    yap.append(y[ind])
    num = num + 1

#print(xap)
start = min(xap)
end = max(xap)
print(start)
print(end)
print(num)

ave = (max(yap)-min(yap))/num
print(ave)


model = np.poly1d(np.polyfit(x, y, 2)) 

polyline = np.linspace(0, 10, 100)
plt.plot(polyline, model(polyline)) 
plt.plot(x, y)
plt.xlim([0, .7])
plt.ylim([-0.21,-0.215])
plt.title('Perpendicular Magnetic Field 1')
plt.xlabel('Positition (m)')
plt.ylabel('Field Strength (G)')
plt.show()
```
